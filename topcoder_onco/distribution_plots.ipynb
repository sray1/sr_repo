```python
# load modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
# load csv file of TS & DX features
df = pd.read_csv('train_feature.csv',index_col=0)
df = df.drop(['CaseID'], axis=1)
print df.shape
df.head()
```

```python
# draw heatmap of all features
corrmap = df.corr()
plt.figure(figsize=(13,13))
ax = sns.heatmap(corrmap,annot=True,fmt=".2f")
plt.show()
```

```python
num_cols = ['f1_DX','f2_DX','f3_DX','f4_DX','f5_DX','f6_DX','f7_DX','f8_DX','f9_DX',
           'f10_DX','f11_DX','f12_DX','f13_DX']
temp_df = df[num_cols]
temp_df['FGA']=df['FGA']

for comb in num_cols:
    #g = sns.FacetGrid(temp_df, size=7)
    g = sns.lmplot(x="FGA", y=comb, data=temp_df, size=7)
    #g.map(plt.scatter,'FGA',comb, alpha=.7)
    plt.show()
```
