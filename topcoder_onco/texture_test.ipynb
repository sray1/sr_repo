```python
from scipy.misc import imread
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import mahotas as mh
import numpy as np
import pandas as pd
import re
from PIL import Image
import time
from skimage.feature import greycomatrix, greycoprops
%matplotlib inline  
```

```python
# load training.csv file with case id and fga numbers
train_df = pd.read_csv('csv/training.csv',header=None)
train_df.head()
```

```python
# load test.csv file with case id and fga numbers
test_df = pd.read_csv('csv/test.csv',header=None)
test_df.head()
```

```python
# from os import listdir
# from os.path import isfile, join
# mypath = "C:\\Users\\Shonket\\Google Drive\\Contest Images\\TS\\"
# onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
# print onlyfiles
# save all pathnames
import glob
train_TS = glob.glob('C:\\Users\\Shonket\\Google Drive\\Contest Images\\TS\\*.png')
#print train_TS
train_DX = glob.glob('C:\\Users\\Shonket\\Google Drive\\Contest Images\\DX\\*.png')
```

```python
train_df_TS = []
start_time = time.time()
# loop through each folder to form feature matrix
for im_filename in train_TS:
    # load and display black and white image
    im = Image.open(im_filename).convert("L")
    arr = np.asarray(im)
    #print arr.shape
    #print type(arr)
    #plt.imshow(arr, cmap = cm.Greys_r)
    #plt.show()
    # find haralick features
    f = mh.features.haralick(arr,return_mean=True)
    #print f
    train_case_id = re.split(r"\\|\.|\-",im_filename)[6]
    #print train_case_id
    data_row = f.tolist()
    data_row.insert(0, train_case_id)
    #print data_row
    train_df_TS.append(data_row) 
    #ind += 1
    
end_time = time.time()
print("Elapsed time was %g seconds" % (end_time - start_time))
```

```python
df_train_TS = pd.DataFrame(train_df_TS)
df_train_TS.columns = ['CaseID','f1_TS','f2_TS','f3_TS','f4_TS','f5_TS','f6_TS','f7_TS','f8_TS',
                      'f9_TS','f10_TS','f11_TS','f12_TS','f13_TS',]
#df_train_TS.columns.values
#df_train_TS = df_train_TS.set_index(['CaseID'])
df_train_TS.head()
df_train_TS.to_csv('TS.csv')
```

```python
df_train_DX = pd.DataFrame(train_df_DX)
df_train_DX.columns = ['CaseID','f1_DX','f2_DX','f3_DX','f4_DX','f5_DX','f6_DX','f7_DX','f8_DX',
                      'f9_DX','f10_DX','f11_DX','f12_DX','f13_DX',]
#df_train_TS.columns.values
#df_train_TS = df_train_TS.set_index(['CaseID'])
df_train_DX.head()
df_train_DX.to_csv('DX.csv')
```

```python
df_train = pd.merge(df_train_DX,df_train_TS,on='CaseID')
df_train.head()
#df_train.describe()
df_train['CaseID'].dtype
df_train['CaseID'] = df_train['CaseID'].astype(long)
df_train['CaseID'].dtype
```

```python
train_df = train_df.rename(columns={0:'CaseID',1:'FGA'})
train_df.head()
train_df['CaseID'].dtype
```

```python
# build feautre matrix dataframe
train_feature = pd.merge(train_df, df_train, on="CaseID", how='left')
train_feature
#train_feature.to_csv('train_feature.csv')
```

```python
X_df = train_feature.drop(['CaseID','FGA'], axis=1)
X_df.head()
x_train = X_df.as_matrix()
x_train
Y_df = train_feature['FGA']
Y_df.head()
y_train = Y_df.as_matrix()
y_train
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=150, min_samples_split=1)
regressor.fit(x_train, y_train)
print regressor.predict(x_train)
```
