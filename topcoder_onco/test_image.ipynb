```python
from scipy.misc import imread
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import mahotas as mh
import numpy as np
import re
from PIL import Image
from skimage.feature import greycomatrix, greycoprops
%matplotlib inline  
```

```python
# load and display image
file_loc = "C:\\Users\\Shonket\\Google Drive\\Contest Images\\TS\\49252-TS.png"
img = imread(file_loc)
print img.shape
print type(img)
plt.imshow(img, cmap=plt.cm.gray)
plt.show()
# load and display black and white image
im = Image.open(file_loc).convert("L")
arr = np.asarray(im)
print arr.shape
print type(arr)
plt.imshow(arr, cmap = cm.Greys_r)
plt.show()
```

```python
# haralick features calc
f = mh.features.haralick(arr,return_mean=True)
#print mh.features.haralick(arr).mean(0)
print f
```

```python
# compute some GLCM properties each patch
xs = []
ys = []
glcm = greycomatrix(arr, [5], [0], 256, symmetric=True, normed=True)
xs.append(greycoprops(glcm, 'dissimilarity')[0, 0])
ys.append(greycoprops(glcm, 'correlation')[0, 0])
print xs
print ys
```
